```python
# ==========================================
# Sentiment Analysis Results Visualization
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "synthetic_sentiment_analysis_results.csv"  # Change path if needed
df = pd.read_csv(file_path)

# Display basic info
print("üîπ First 5 rows:")
print(df.head(), "\n")

print("üîπ Dataset info:")
print(df.info(), "\n")

# Check for missing values
print("üîπ Missing values per column:")
print(df.isnull().sum(), "\n")

# Drop missing or duplicate entries
df = df.dropna().drop_duplicates()

# Automatically detect sentiment column
sentiment_col = None
for col in df.columns:
    if "sentiment" in col.lower():
        sentiment_col = col
        break
if not sentiment_col:
    sentiment_col = df.columns[-1]  # fallback to last column
print(f"‚úÖ Using sentiment column: {sentiment_col}\n")

# Sentiment distribution summary
print("üîπ Sentiment counts:")
print(df[sentiment_col].value_counts(), "\n")

# Bar chart
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x=sentiment_col)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment Category")
plt.ylabel("Count")
plt.show()

# Pie chart
plt.figure(figsize=(6, 6))
df[sentiment_col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Sentiment Percentage Distribution")
plt.ylabel("")
plt.show()

print("üéØ Visualization complete!")
```
